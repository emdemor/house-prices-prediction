```python
class FeatureRemover(BaseEstimator,TransformerMixin):
    
    def __init__(self,features,active = True):
        self.active = active
        self.features = features
        
    def fit(self,X,y=None):
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            self.X = X.copy() 
            if type(self.features) == str:
                return self.X.drop(columns=[self.features])
            elif type(self.features) == list:
                return self.X.drop(columns = self.features)

class FeatureSelector(BaseEstimator,TransformerMixin):
    
    def __init__(self,features,active = True):
        self.features = features
        self.active = active
        
    def fit(self,X,y=None):
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            self.X = X.copy() 
            return self.X[self.features]
            
```

```python
reset_datasets()

class SalePriceTransformer(BaseEstimator,TransformerMixin):
    
    def __init__(self,active=True,target_name = 'SalePrice'):
        self.target_name = target_name
        self.active = active
        
    def fit(self,y):
        self.log_ymin = np.log10(y.min())
        self.log_ymax = np.log10(y.max())
        return self
        
    def transform(self,y):
        if not self.active:
            return y
        else:
            return self.__transformation(y)
        
    def __transformation(self,y_in):
        y = y_in.copy()
        log_y = np.log10(y)
        return log_y
    
    def inverse_transform(self,y):
        if not self.active:
            return y
        else:
            return self.__inv_transformation(y)
        
    def __inv_transformation(self,log_y_in):
        
        log_y = log_y_in.copy()
        
        y = 10**(log_y)
        
        return y.astype(int)
        
            
TargetTransformer = SalePriceTransformer()
TargetTransformer.fit(y_trn)

y_trn = TargetTransformer.transform(y_trn)
y_trn
#TargetTransformer.inverse_transform(y_trn)
```

```python
# class LandContourTransformer(BaseEstimator,TransformerMixin):
class TargetBasedFeatureTransformer(BaseEstimator,TransformerMixin):
    def __init__(self,feature_name,target_name = 'SalePrice',active=True,n_class = 3):
        self.feature_name = feature_name
        self.target_name = target_name
        self.active = active
        self.n_class = n_class
        
    def fit(self,X,y=None):
        
        df = pd.concat([X,y],axis=1)
        
        self.mode = X[self.feature_name].mode().values[0]

        mean_target = df[[self.feature_name,self.target_name ]].groupby(self.feature_name).agg(['mean'])
        max_target = mean_target.iloc[:,0].max()
        min_target = mean_target.iloc[:,0].min()
        amp_class = (max_target-min_target)/self.n_class
        limits = [min_target + i*amp_class for i in range(0,self.n_class+1)]

        def evaluate_target(num_teste,n_class,limits):
            for i in range(0,n_class):

                if num_teste < limits[0]:
                    return 0

                if num_teste > limits[-1]:
                    return n_class+1

                if num_teste >= limits[i] and num_teste <= limits[i+1]:
                    return i+1

        self.dict_subs = {}
        for index, row in mean_target.iterrows():
            self.dict_subs[index] = evaluate_target(row[self.target_name].values[0],self.n_class,limits)
            
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            return self.__transformation(X)
        
    def __transformation(self,X_in):
        X = X_in.copy()
        X[self.feature_name] = X[self.feature_name].fillna(self.mode)
        new_list = []
        for index, row in X.iterrows():
            new_list.append(self.dict_subs[row[self.feature_name]])
                
        X[self.feature_name] = new_list
            
        return X
```

```python
class MSSubClassTransformer(BaseEstimator,TransformerMixin):
    
    def __init__(self,active=True,feature_name = 'MSSubClass',target_name = 'SalePrice'):
        self.feature_name = feature_name
        self.target_name = target_name
        self.active = active
        
    def fit(self,X,y=None):
        self.X = X.copy()
        self.y = y.copy()
        self.df = pd.concat([X,y],axis = 1)
        
        MSSubClass_prices = self.df[[self.feature_name,self.target_name]].groupby(by = self.feature_name).agg(['mean'])
        MSSubClass_prices['price_norm'] = MSSubClass_prices[(self.target_name, 'mean')]
        MSSubClass_prices = MSSubClass_prices.sort_values(by='price_norm',ascending = False)
        
        PricesMean = self.df[self.target_name].mean()
        
        self.dict_transform = {}
        self.dict_inv_transform = {}
       
        for index, row in MSSubClass_prices.iterrows():
            val = int(round(100*row['price_norm'].values[0]/PricesMean))
            self.dict_transform[index] = val
            self.dict_inv_transform[val] = index
            
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            return self.__transformation(X)
        
    def __transformation(self,X_in):
        X = X_in.replace(150,160).copy()
        X[self.feature_name] = [self.dict_transform[elem] for elem in np.array(X[self.feature_name])]
        return X
            
```

```python
class MSZoningTransformer(BaseEstimator,TransformerMixin):
    
    def __init__(self,active=True,feature_name = 'MSZoning',target_name = 'SalePrice'):
        self.feature_name = feature_name
        self.target_name = target_name
        self.active= active
        
    def fit(self,X,y=None):
        
        self.X = X.copy()
        self.y = y.copy()
        self.df = pd.concat([X,y],axis = 1)
        
        MSSubClass_prices = self.df[[self.feature_name,self.target_name]].groupby(by = self.feature_name).agg(['mean'])
        MSSubClass_prices['price_norm'] = MSSubClass_prices[(self.target_name, 'mean')]
        MSSubClass_prices = MSSubClass_prices.sort_values(by='price_norm',ascending = False)
        
        PricesMean = self.df[self.target_name].mean()
        
        self.dict_transform = {}
        self.dict_inv_transform = {}
        ind = 0 
        for index, row in MSSubClass_prices.iterrows():
            #ind += 1
            val = ind
            val = int(round(100*row['price_norm'].values[0]/PricesMean))
            self.dict_transform[index] = val
            self.dict_inv_transform[val] = index
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            return self.__transformation(X)
        
    def __transformation(self,X_in):
        X = X_in.copy()
        md = self.X[self.feature_name].mode().values[0]
        X[self.feature_name] = X[self.feature_name].fillna(md)
        X[self.feature_name] = [self.dict_transform[elem] for elem in np.array(X[self.feature_name])]
        return X
            
```

```python
class LotShapeTransformer(BaseEstimator,TransformerMixin):
    
    def __init__(self,active=True,feature_name = 'LotShape',target_name = 'SalePrice',factor = 100):
        self.feature_name = feature_name
        self.target_name = target_name
        self.active = active
        self.factor = factor
        
    def fit(self,X,y=None):
        self.X = X.copy()
        self.y = y.copy()
        self.df = pd.concat([X,y],axis = 1)
        
        LotShape_prices = self.df[[self.feature_name,self.target_name]].groupby(by = self.feature_name).agg(['mean'])
        LotShape_prices['price_norm'] = LotShape_prices[(self.target_name, 'mean')]
        LotShape_prices = LotShape_prices.sort_values(by='price_norm',ascending = False)
        
        PricesMean = self.df[self.target_name].mean()
        
        self.dict_transform = {}
        self.dict_inv_transform = {}
        
        val_list = [ int(round(self.factor*row['price_norm'].values[0]/PricesMean)) for index, row in LotShape_prices.iterrows()]
        min_val = np.min(val_list)
        max_val = np.max(val_list)

        for index, row in LotShape_prices.iterrows():
            val = int(round(self.factor*row['price_norm'].values[0]/PricesMean))
            self.dict_transform[index] = int(self.factor*(val - min_val)/(max_val - min_val))
            self.dict_inv_transform[val] = index
            
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            return self.__transformation(X)
        
    def __transformation(self,X_in):
        X = X_in.replace(150,160).copy()
        X[self.feature_name] = [self.dict_transform[elem] for elem in np.array(X[self.feature_name])]
        return X
            
```

```python
# class LandContourTransformer(BaseEstimator,TransformerMixin):
class LandContourTransformer(BaseEstimator,TransformerMixin):
    
    def __init__(self,active=True,feature_name = 'LandContour',target_name = 'SalePrice',factor = 100):
        self.feature_name = feature_name
        self.target_name = target_name
        self.active = active
        self.factor = factor
        
    def fit(self,X,y=None):
        self.X = X.copy()
        self.y = y.copy()
        self.df = pd.concat([X,y],axis = 1)
        
        LotShape_prices = self.df[[self.feature_name,self.target_name]].groupby(by = self.feature_name).agg(['mean'])
        LotShape_prices['price_norm'] = LotShape_prices[(self.target_name, 'mean')]
        LotShape_prices = LotShape_prices.sort_values(by='price_norm',ascending = False)
        
        PricesMean = self.df[self.target_name].mean()
        
        self.dict_transform = {}
        self.dict_inv_transform = {}
        
        val_list = [ int(round(self.factor*row['price_norm'].values[0]/PricesMean)) for index, row in LotShape_prices.iterrows()]
        min_val = np.min(val_list)
        max_val = np.max(val_list)

        for index, row in LotShape_prices.iterrows():
            val = int(round(self.factor*row['price_norm'].values[0]/PricesMean))
            self.dict_transform[index] = int(self.factor*(val - min_val)/(max_val - min_val))
            self.dict_inv_transform[val] = index
            
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            return self.__transformation(X)
        
    def __transformation(self,X_in):
        X = X_in.copy()
        X[self.feature_name] = [self.dict_transform[elem] for elem in np.array(X[self.feature_name])]
        return X
```

```python
# class LandContourTransformer(BaseEstimator,TransformerMixin):
class NeighborhoodTransformer(BaseEstimator,TransformerMixin):
    
    def __init__(self,active=True,feature_name = 'Neighborhood',target_name = 'SalePrice',n_class = 3):
        self.feature_name = feature_name
        self.target_name = target_name
        self.active = active
        self.n_class = n_class
        
    def fit(self,X,y=None):
        
        df = pd.concat([X,y],axis=1)

        mean_target = df[[self.feature_name,self.target_name ]].groupby(self.feature_name).agg(['mean'])
        max_target = mean_target.iloc[:,0].max()
        min_target = mean_target.iloc[:,0].min()
        amp_class = (max_target-min_target)/self.n_class
        limits = [min_target + i*amp_class for i in range(0,self.n_class+1)]

        def evaluate_target(num_teste,n_class,limits):
            for i in range(0,n_class):

                if num_teste < limits[0]:
                    return 0

                if num_teste > limits[-1]:
                    return n_class+1

                if num_teste >= limits[i] and num_teste <= limits[i+1]:
                    return i+1

        self.dict_subs = {}
        for index, row in mean_target.iterrows():
            self.dict_subs[index] = evaluate_target(row['SalePrice'].values[0],self.n_class,limits)
            
        return self
        
    def transform(self,X):
        if not self.active:
            return X
        else:
            return self.__transformation(X)
        
    def __transformation(self,X_in):
        X = X_in.copy()
        
        new_list = []
        for index, row in X.iterrows():
            new_list.append(self.dict_subs[row[self.feature_name]])
                
        X[self.feature_name] = new_list
            
        return X
```
